```python
def _parse_key(key: str):
    components = key.split(".")
    parsed_components = []
    for i in range(len(components)):
        parsed_components.append(
            ".".join(components[:i]) + ("." if i != 0 else "") + components[i]
        )
    return parsed_components
```
